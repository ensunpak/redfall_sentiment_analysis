```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import datetime as dt
import pickle
```

```python
# Define date range to scrape the data. Scrape 100 tweets per day
scrape_range = pd.date_range(start=dt.datetime(2023, 2, 2), end=dt.datetime(2023, 6, 1), freq="D")
```

```python
# Initialize list to store the tweets
tweets = []

# Setup the scraping procedure by hashtag
for dte in scrape_range:
    mth_tweets = []
    results = sntwitter.TwitterHashtagScraper("Redfall since:" + dte.strftime("%Y-%m-%d") + " until:" + (dte + pd.Timedelta(1, "D")).strftime("%Y-%m-%d")).get_items()
    
    for i, tweet in enumerate(results):
        if i >= 100:
            break
        mth_tweets.append([tweet.date, tweet.user.username, tweet.lang, tweet.rawContent])
    [tweets.append(twt) for twt in mth_tweets]
```

```python
# Download tweets into a pickle file
with open("redfall_tweets_hash.pkl", "wb") as f:
    pickle.dump(tweets, f)
f.close()
```
